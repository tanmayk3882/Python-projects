```python
import tkinter as tk
from tkinter import messagebox
import os
import csv
file_path = "D:/StudentData/student_data.csv"
os.makedirs(os.path.dirname(file_path), exist_ok=True)
if not os.path.isfile(file_path):
    with open(file_path, mode='w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(["Name", "Roll No", "Enrollment No", "Branch", "College", "Project Name", "Marks"])
def save_data():
    name = name_entry.get()
    roll = roll_entry.get()
    enroll = enroll_entry.get()
    branch = branch_entry.get()
    college = college_entry.get()
    project = project_entry.get()
    marks = marks_entry.get()
    if not all([name, roll, enroll, branch, college, project, marks]):
        messagebox.showerror("Error", "Please fill all fields.")
        return
    with open(file_path, mode='a', newline='') as file:
        writer = csv.writer(file)
        writer.writerow([name, roll, enroll, branch, college, project, marks])
    name_entry.delete(0, tk.END)
    roll_entry.delete(0, tk.END)
    enroll_entry.delete(0, tk.END)
    branch_entry.delete(0, tk.END)
    college_entry.delete(0, tk.END)
    project_entry.delete(0, tk.END)
    marks_entry.delete(0, tk.END)
    messagebox.showinfo("Data saved successfully!")
window = tk.Tk()
window.title("Student Data Entry")
window.geometry("400x500")
tk.Label(window, text="Name:").pack()
name_entry = tk.Entry(window, width=45)
name_entry.pack()
tk.Label(window, text="Roll No:").pack()
roll_entry = tk.Entry(window, width=45)
roll_entry.pack()
tk.Label(window, text="Enrollment No:").pack()
enroll_entry = tk.Entry(window, width=45)
enroll_entry.pack()
tk.Label(window, text="Branch:").pack()
branch_entry = tk.Entry(window, width=45)
branch_entry.pack()
tk.Label(window, text="College:").pack()
college_entry = tk.Entry(window, width=45)
college_entry.pack()
tk.Label(window, text="Project Name:").pack()
project_entry = tk.Entry(window, width=45)
project_entry.pack()
tk.Label(window, text="Marks:").pack()
marks_entry = tk.Entry(window, width=45)
marks_entry.pack()
tk.Button(window, text="Save", command=save_data, bg="green", fg="white", width=10).pack(pady=30)
window.mainloop()
```
